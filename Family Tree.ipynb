```python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender
        self.spouse = None
        self.children = []
        self.mother = None
        self.father = None
```

```python
class FamilyTree:
    def __init__(self):
        self.members = {}

    def add_member(self, person):
        self.members[person.name] = person

    def add_spouse(self, name1, name2, gender2):
        if name1 not in self.members:
            return

        person1 = self.members[name1]
        person2 = Person(name2, gender2)

        person1.spouse = person2
        person2.spouse = person1

        self.members[name2] = person2

    def add_child(self, mother_name, child_name, gender):
        if mother_name not in self.members:
            return "PERSON_NOT_FOUND"

        mother = self.members[mother_name]

        child = Person(child_name, gender)
        child.mother = mother

        if mother.spouse:
            child.father = mother.spouse

        mother.children.append(child)
        self.members[child_name] = child

        return "CHILD_ADDITION_SUCCEEDED"

    def get_relationship(self, name, relation):
        if name not in self.members:
            return ["PERSON_NOT_FOUND"]

        person = self.members[name]

        if relation == "Son":
            return [c.name for c in person.children if c.gender == "Male"] or ["NONE"]

        elif relation == "Daughter":
            return [c.name for c in person.children if c.gender == "Female"] or ["NONE"]

        elif relation == "Siblings":
            if not person.mother:
                return ["NONE"]
            return [
                c.name for c in person.mother.children
                if c.name != person.name
            ] or ["NONE"]

        elif relation == "Paternal-Uncle":
            if not person.father or not person.father.mother:
                return ["NONE"]
            return [
                c.name for c in person.father.mother.children
                if c.gender == "Male" and c.name != person.father.name
            ] or ["NONE"]

        elif relation == "Maternal-Uncle":
            if not person.mother or not person.mother.mother:
                return ["NONE"]
            return [
                c.name for c in person.mother.mother.children
                if c.gender == "Male" and c.name != person.mother.name
            ] or ["NONE"]

        return ["NONE"]
```

```python
def setup_family():
    family = FamilyTree()

    arthur = Person("Arthur", "Male")
    margaret = Person("Margaret", "Female")

    arthur.spouse = margaret
    margaret.spouse = arthur

    family.add_member(arthur)
    family.add_member(margaret)

    # Children
    family.add_child("Margaret", "Bill", "Male")
    family.add_child("Margaret", "Charlie", "Male")
    family.add_child("Margaret", "Percy", "Male")
    family.add_child("Margaret", "Ron", "Male")
    family.add_child("Margaret", "Ginny", "Female")

    # Spouses
    family.add_spouse("Bill", "Flora", "Female")
    family.add_spouse("Charlie", "Audrey", "Female")
    family.add_spouse("Percy", "Victoria", "Female")
    family.add_spouse("Ron", "Helena", "Female")
    family.add_spouse("Ginny", "Harry", "Male")

    # Next Generation
    family.add_child("Flora", "Victoire", "Female")
    family.add_child("Flora", "Dominique", "Female")
    family.add_child("Flora", "Louis", "Male")

    return family
```

```python
def process_commands(commands, family):
    for command in commands:
        parts = command.strip().split()

        if parts[0] == "ADD_CHILD":
            result = family.add_child(parts[1], parts[2], parts[3])
            print(result)

        elif parts[0] == "GET_RELATIONSHIP":
            result = family.get_relationship(parts[1], parts[2])
            print(" ".join(result))
```

```python
family = setup_family()

commands = [
    "ADD_CHILD Flora Minerva Female",
    "GET_RELATIONSHIP Minerva Siblings",
    "GET_RELATIONSHIP Bill Son",
    "GET_RELATIONSHIP Ginny Paternal-Uncle",
    "GET_RELATIONSHIP Harry Maternal-Uncle"
]

process_commands(commands, family)
```
